```python
def fac(n):
    x = 1
    while n!=1:
        x = x*n
        n = n-1
    return x
fac(7)
        
```

```python
def fact(n):
    if n==0 or n==1:
        return n
    else:
        return fact(n)*fact(n-1)
fact(5)
```

```python
def age_calculator():
    c_date = input("ENTER THE CURRENT DATE IN format of ddmmyyyy:")
    b_date = input("ENTER THE BIRTH DATE IN format of ddmmyyyy:")
    print(c_date)
    print(b_date)
    dod =int(b_date[0:2])
    dom =int(b_date[2:4])
    doy =int(b_date[4:])
    cod =int(b_date[0:2])
    com =int(b_date[2:4])
    coy =int(b_date[4:])
     
    birth_days = 365*doy + 30*dom + dod
    current_days = 365*coy + 30*com + cod
    age = current_days - birth_days

    birth_year = age//365
    reminder = age%365
    birth_month = reminder//30
    birth_date = reminder%30
    
   
    print("you are",birth_year,"years",birth_month,"months",birth_date,"days")
    

age_calculator()
```
